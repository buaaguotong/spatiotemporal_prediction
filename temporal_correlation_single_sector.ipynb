```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
plt.rcParams['font.sans-serif']=['Microsoft YaHei']
```

```python
data_path = 'gd_data/newdata/factors/sector1.csv'
df = pd.read_csv(data_path, header=None)
```

```python
df.columns
```

```python
corr = abs(df.corr().fillna(0))

plt.figure(figsize=[10,10])
plt.title('Cross-correlation')
ax = sns.heatmap(
    corr, 
#     cmap="vlag",
    square=True
)
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf


for col in df.columns[:24]:
    fig = plt.figure(figsize=[18,2])
    ax1 = fig.add_subplot(121)
    ax2 = fig.add_subplot(122)
    fig.suptitle(tag[int(col)])
    plot_acf(df[col].values.squeeze(), ax1, lags=50)
    plot_pacf(df[col].values.squeeze(), ax2, lags=50)
    
    plt.show()
```
